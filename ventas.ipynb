```python
query = """ 
SELECT Vendedor, Producto, SUM(Ventas) TVentas
FROM df2 
WHERE FPago like '%Tarjeta%'
GROUP BY Vendedor, Producto
ORDER BY Vendedor, Producto

"""
```

```python
q1=sqldf(query)
q1
```

```python
query = """ 
SELECT Ciudad,  SUM(Ventas) TVentas
FROM df2 
--WHERE FPago like '%Tarjeta%'
GROUP BY Ciudad
ORDER BY TVentas Desc

"""
```

```python
query = """ 
SELECT Vendedor,  SUM(Ventas) TVentas
FROM df2 
--WHERE FPago like '%Tarjeta%'
GROUP BY Vendedor
ORDER BY TVentas Desc

"""
q1=sqldf(query)
#q1.set_index('Vendedor', inplace=True)
plt.barh(q1['Vendedor'], q1['TVentas'], color="#998844", label="Vendedor")
plt.legend()
plt.title("Ventas")
plt.xlabel("Monto ($)")
plt.ylabel("Vendedor")
#q1.plot(kind='barh')
plt.show()
```

```python
query = """ 
SELECT FPago,  SUM(Ventas) TVentas
FROM df2 
--WHERE FPago like '%Tarjeta%'
GROUP BY FPago
ORDER BY TVentas Desc

"""
q1=sqldf(query)
q1
```

```python
slices = q1['TVentas']
labels = q1['FPago']
plt.style.use("fivethirtyeight")
explode = [0.1,0,0]
plt.pie(slices, labels=labels, 
        explode=explode, 
        wedgeprops={'edgecolor':'black'}, 
        shadow=True,
        startangle= -90,
        autopct='%1.1f%%'
        )

plt.title("Ventas")
plt.show()
```

```python
query = """ 
SELECT Producto,  SUM(Ventas) TVentas
FROM df2 
GROUP BY Producto 
ORDER BY TVentas Desc

"""
q1=sqldf(query)
#q1.set_index('Vendedor', inplace=True)
plt.barh(q1['Producto'], q1['TVentas'], color="#cc4499", label="Producto")
plt.legend()
plt.title("Ventas")
plt.xlabel("Monto ($)")
plt.ylabel("Vendedor")
#q1.plot(kind='barh')
plt.show()
```
