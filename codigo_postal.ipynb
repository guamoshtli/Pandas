```python
# 1. Distribution of sex and their relation credit mount
sex_types = df2['SEXO'].value_counts()
sex_types
```

```python
sex_mount = df2.groupby('SEXO')['MONTO_MAX_AUT'].mean()
sex_mount
```

```python
# Format labels for the second graph
formatted_sex_mount = ['$' + f'{value:,.2f}' for value in sex_mount.values]
formatted_sex_mount
```

```python
plt.bar(sex_types, sex_mount, color="#998844", label="Sex")
plt.legend()
plt.title("Sex")
plt.ylabel("Monto ($)")
plt.xlabel("Sex")
#q1.plot(kind='barh')
plt.show()
```

```python
query = """ 
SELECT SEXO,  SUM(MONTO_MAX_AUT) MOtorgado
FROM df2 

GROUP BY SEXO
ORDER BY MOtorgado Desc

"""
q1=sqldf(query)
#q1.set_index('Vendedor', inplace=True)
plt.bar(q1['SEXO'], q1['MOtorgado'], color=[ 'blue', 'pink'], label="SEXO")
#plt.legend()
plt.title("SEXO")
plt.ylabel("Monto ($)")
plt.xlabel("SEXO")
#q1.plot(kind='barh')
plt.show()
```

```python
query = """ 
SELECT DELEGACIÓN_O_MUNICIPIO DEL,  COUNT(MONTO_MAX_AUT) NUM_CREDS
FROM df2 

GROUP BY DELEGACIÓN_O_MUNICIPIO
ORDER BY NUM_CREDS ASC

"""
q1=sqldf(query)
#q1.set_index('Vendedor', inplace=True)
plt.barh(q1['DEL'], q1['NUM_CREDS'], color=[ 'blue', 'black', 'green', 'purple'], label="MUNICIPIO")
#plt.legend()
plt.title("CREDITOS POR MUNICIPIO")
plt.xlabel("NUM. CREDITOS")
plt.ylabel("MUNICIPIO")
#q1.plot(kind='barh')
plt.show()
```
