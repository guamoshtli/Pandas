```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python=country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_pd = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_pd
```

```python
python_pd.rename(columns={'Country': 'Total', 'LanguageWorkedWith':'Python_Tot'}, inplace=True)
```

```python
python_pd
```

```python
python_pd['PythonPct']= python_pd['Python_Tot']*100/python_pd['Total']
```

```python
python_pd.nlargest(columns='Total',n=20)
```
