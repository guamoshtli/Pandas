#  Variables are stored in both rows and columns

```python
import datetime

df_weather = pd.read_csv('weather-raw.csv')

df_weather.sample(10)
```

```python
# Melting

df_weather= df_weather.melt(id_vars=['id', 'year', 'month', 'element'],var_name=["day"],value_name="temp")
df_weather
```

```python
#Cleaning data 

df_weather['day']= df_weather['day'].apply(lambda x : x.replace('d',''))
```

```python
df_weather.sample(10)
```

```python
# applying pivot_table to create columns for tmin and tmax
df_weather= df_weather.pivot_table(index=['id', 'year', 'month', 'day'],  columns='element', values='temp' )
df_weather
```

```python
#Reset index
df_weather= df_weather.reset_index()
df_weather
```

```python
df_weather.info()
```

```python
# Crating a date column

df_weather['Date'] = df_weather['year'].astype('str')+'/'+df_weather['month'].astype('str')+'/'+df_weather['day']
df_weather
```

```python
df_weather.info()
```

```python
df_weather['Date'] =  pd.to_datetime(df_weather['Date'], format='%Y/%m/%d')
```

```python
df_weather.info()
```

```python
# Grouping by date
df_weather_grp = df_weather.groupby('Date')
```

```python
df_weather_grp['tmax', 'tmin'].mean()
```

# Multiple types of observational units are stored in the same table.

```python
# Importing the dataset
import pandas as pd
import re
import numpy as np
import datetime

df_bill = pd.read_csv('billboard.csv',header=0,encoding = 'unicode_escape')

df_bill.head()
```

```python
df_bill.shape
```

```python
df_bill.info()
```

```python
df_bill.columns
```

```python
# Applying melt

df_bill= df_bill.melt(id_vars=['year', 'artist.inverted', 'track', 'time', 'genre', 'date.entered',
       'date.peaked'], var_name='week', value_name='rank')
df_bill
```

```python
# Week to number 
df_bill['week'].unique()
```

```python
df_bill['week']=df_bill['week'].str.replace(r'[a-zA-Z.,\s]+', '', regex=True)
df_bill['week'].unique()
```

```python
df_bill
```

```python
df_bill.info()
```

```python
df_bill['week']=df_bill['week'].astype('int')
```

```python
df_bill.describe()
```

```python
# creating a date column from date.entered and week

df_bill['date.entered'] =  pd.to_datetime(df_bill['date.entered'], format='%Y/%m/%d')
```

```python

df_bill.info()
```

```python
df_bill= df_bill.rename(columns={'date.entered':'date_entered', 'date.peaked':'date_peaked'})
df_bill.head()
```

```python
df_bill['date_entered']+datetime.timedelta(weeks= 1)
```

```python
df_bill['date']= df_bill.apply(lambda x:  x['date_entered'] + datetime.timedelta(weeks=x['week']-1), axis=1)
```

```python
df_bill.head(10)
```

```python
# Get away from unecesary data

df_bill= df_bill.drop(['date_entered','date_peaked','week'], axis=1)\
.sort_values('artist.inverted', ascending=True)\
.reset_index(drop=True)

df_bill.head()
```

```python
df_bill['rank'].unique()
```
